```python
# Importing data libraries in python
import sklearn.datasets as datasets
import pandas as pd
from sklearn.metrics import mean_absolute_error

# Loading the iris datasets
iris = datasets.load_iris()

# Forming the iris DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(10))

# Setting up the varible we want to fit
y = iris.target
print(y)
```

# **Defining the Decision Tree Algorithm**

```python
# Defining decision tree algorithm
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
# Fitting the data
dtree.fit(df, y)

# Predicting the fitted values and storing them ins iris_predict
iris_predict = dtree.predict(df)
print(iris_predict)
print('Mean Absolute Error is', mean_absolute_error(y, iris_predict))
print('Decision Tree Classifer Created')
```

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
